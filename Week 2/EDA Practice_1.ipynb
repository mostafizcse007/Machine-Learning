```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
df.drop(columns=['Unnamed: 32'],inplace=True)
```

```python
df.describe().T
```

```python
df['diagnosis'].value_counts().plot(kind='bar')
plt.show()
```

```python
numeric = []
for c in df.columns:
    if df[c].dtype != 'object':
        numeric.append(c)
numeric
```

```python
df[numeric].agg(['mean','median','std','min','max']).T
```

```python
df[numeric].hist(bins=10,figsize=(18,8))
plt.tight_layout()
plt.show()
```

```python
for col in numeric:
    plt.figure(figsize=(8,4))
    sns.histplot(data=df,x=col,bins=10,kde = True)
    plt.tight_layout()
    plt.show()
```

```python
df[numeric].boxplot(figsize=(16,8))
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```
