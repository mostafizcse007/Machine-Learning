##### Euclidean vs Manhattan

```python
x = np.array([
    [70,80],
    [60,90],
    [85,60],
    [78,76],
    [62,65]
],dtype=float)
q = np.array([75,70],dtype=float)
```

```python
# euclidean distance(L2)
eu = np.linalg.norm(x-q, axis=1)
print('Euclidean distance',np.round(eu,3).tolist())
```

```python
# manhattan distance(L1)
man = np.linalg.norm(x-q, axis=1,ord = 1)
print('Manhattan distance',np.round(man,3).tolist())
```
