```python
weights = np.array([2.5, 3.0, 3.2, 3.3, 3.4, 3.5, 3.6, 3.9, 4.0, 4.5])
```

```python
q1 = np.percentile(weights,25)
q2 = np.percentile(weights,50)
q3 = np.percentile(weights,75)
lower_bound = q1 - 1.5*(q3-q1)
upper_bound = q3 + 1.5*(q3-q1)
outlires = weights[(weights<lower_bound) | (weights>upper_bound)]
outlires
```
