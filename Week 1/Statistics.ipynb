```python
import numpy as np
```

```python
salaries = np.array([45, 48, 50, 52, 55, 58, 60, 62, 65, 68, 70, 72, 75, 78, 80, 85, 90, 95, 150, 500])
```

```python
sal_range = np.ptp(salaries)
variance = np.var(salaries, ddof = 1)
std = np.std(salaries, ddof= 1)
q1 = np.percentile(salaries, 25)
q2 = np.percentile(salaries, 50)
q3 = np.percentile(salaries, 75)
iqr = q3 - q1
lower_fence = q1 - 1.5*iqr
upper_fence = q3 + 1.5*iqr
outlires = salaries[(salaries<lower_fence) | (salaries>upper_fence)]
```

```python
print('Salary range:', sal_range)
print('Variance:', variance)
print('Standard Deviation:', std)
print(f"Q1: {q1} Q2: {q2} Q3: {q3}")
print('IQR:', iqr)
print(f"Lower fence: {lower_fence} Upper fence: {upper_fence}")
print('Number of outlires:',len(outlires))
print('Outlires',outlires)
```
