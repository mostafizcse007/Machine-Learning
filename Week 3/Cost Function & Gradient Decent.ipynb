```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression,SGDRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score,root_mean_squared_error
```

```python
hours = np.array([2,4,6,8,10]).reshape(-1,1)
marks = np.array([55,60,70,80,88]).reshape(-1,1)
hours,marks
```

###### Fit a Linear Regression model

```python
lr = LinearRegression()
lr.fit(hours,marks)
```

```python
lr.coef_,lr.intercept_
```

###### Compute MSE

```python
predict_lr = lr.predict(hours)
mse_lr = mean_squared_error(marks,predict_lr)
mse_lr
```

##### Visualization

```python
plt.scatter(hours,marks,color='g',label='Actual Marks')
plt.plot(hours,predict_lr,color='r',label='Best fit line')
plt.legend()
plt.tight_layout()
plt.xlabel('Hours studied')
plt.ylabel('Marks')
plt.title('Linear Regression (Study Hours vs Marks)')
plt.show()
```

###### Use a gradient descent based model

```python
sgd = SGDRegressor(
    max_iter=1000,learning_rate='invscaling',eta0=.01,random_state=42
)
sgd.fit(hours,marks)
sgd.coef_,sgd.intercept_
```

```python
predict_sgd = sgd.predict(hours)
mse_sgd = mean_squared_error(marks,predict_sgd)
mse_sgd
```

##### Compare both lines

```python
plt.scatter(hours,marks,color='g',label='Actual Marks')
plt.plot(hours,predict_lr,color='r',label='Linear Regression line',linestyle='solid')
plt.plot(hours,predict_sgd,color='b',label='Gradient Descent line',linestyle=':')
plt.legend()
plt.tight_layout()
plt.xlabel('Hours studied')
plt.ylabel('Marks')
plt.title('Linear Regression (Study Hours vs Marks)')
plt.show()
```

##### Compute metrices

```python
mae = mean_absolute_error(marks,predict_lr)
mse = mean_squared_error(marks,predict_lr)
r2 = r2_score(marks,predict_lr)
rmse = root_mean_squared_error(marks,predict_lr)
mae,mse,r2,rmse
```
