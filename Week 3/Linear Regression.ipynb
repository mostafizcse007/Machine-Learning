```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import seaborn as sns
```

```python
df = pd.read_csv('hours_studied.csv')
```

```python
X_train,X_test,y_train,y_test = train_test_split(df[['hours_studied']],df[['marks']],test_size=.2,random_state=42)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
r2 = r2_score(y_test,y_pred)
r2
```

```python
plt.grid(True)
sns.scatterplot(data=df,x='hours_studied',y='marks',label='Actual Data',marker='o',color='green')
plt.plot(X_test,y_pred,color = 'red',label='Best  fit line')
plt.xlabel('Hours studied')
plt.ylabel('Marks')
plt.title('Study hours vs Marks for Linear Regression')
plt.legend()
plt.tight_layout()
plt.show()
```
